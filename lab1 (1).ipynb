```python
import numpy as np
import pandas as pd
roll_no = np.arange(1, 11)
marks = np.array([78, 85, 90, 66, 72, 88, 91, 69, 75, 80])
attendance = np.array([85, 90, 95, 70, 80, 92, 96, 75, 88, 90])
gender = np.array(['M', 'F', 'M', None, 'F', 'M', None, 'F', 'M', 'F'])
student_df = pd.DataFrame({
    'Roll No': roll_no,
    'Marks': marks,
    'Attendance': attendance,
    'Gender': gender
})
print(student_df)
```

```python

null_count = student_df['Gender'].isnull().sum()
print("Null values in Gender column:", null_count)
mode_gender = student_df['Gender'].mode()[0]
student_df['Gender'].fillna(mode_gender, inplace=True)
print("\nAfter replacing null values:")
print(student_df)
```

```python

average_marks = student_df['Marks'].mean()
print("\nAverage Marks:", average_marks)
gender_count = student_df['Gender'].value_counts()
print("\nCount of Males and Females:")
print(gender_count)
```

```python
import matplotlib.pyplot as plt

gender_count = student_df['Gender'].value_counts()

plt.bar(gender_count.index, gender_count.values)
plt.xlabel("Gender")
plt.ylabel("Number of Students")
plt.title("Gender Distribution of Students")
plt.legend(["Students"])
plt.show()
```

```python
plt.hist(student_df['Marks'])
plt.xlabel("Marks")
plt.ylabel("Number of Students")
plt.title("Marks Distribution of Students")
plt.legend(["Marks"])
plt.show()
```
