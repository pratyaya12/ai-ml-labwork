```python
import numpy as np
import pandas as pd
import random

r_roll=np.random.randint(1,101,size=60)
roll=list(r_roll)

r_height=np.random.randint(140,190,size=60)
height=list(r_height)

r_weight=np.random.randint(25,200,size=60)
weight=list(r_weight)

r_stipend=np.random.randint(5000,75001,size=60)
stipend=list(r_stipend)

name=["ram","rajesh","rahul","virat","rakesh","ramu"]*10

data={"roll":roll,
      "height":height,
      "name":name,
      "weight":weight,
      "stipend":stipend}

index=[]
for i in range(1,61):
    index.append(i)

df=pd.DataFrame(data,index=index)
print(df)

```

```python
# Height normalization

df["min_max_height"] = (df["height"]-df["height"].min())/(df["height"].max() - df["height"].min())

# Weight normalization

df["min_max_weight"] = (df["weight"]-df["weight"].min())/(df["weight"].max() - df["weight"].min())

print(df)
```
